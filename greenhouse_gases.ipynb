### deep analysis


#### **Carbon dioxide**
carbon dioxide sticks around for quite a while. Once it’s emitted into the atmosphere, **40 percent** still remains after **100 years**, 20 percent after 1,000 years, and 10 percent as long as 10,000 years later.


### **Methane**
Methane (CH4) persists in the atmosphere for around **12 years**, which is less time than carbon dioxide, but it is much more potent in terms of the greenhouse effect. In fact, pound for pound, its global warming impact is almost **21 times** greater than that of carbon dioxide over a 100-year period.


### **Nitrous oxide**
Nitrous oxide (N2O) is a powerful greenhouse gas: According to the EPA, it has a GWP that is around **310 times** that of carbon dioxide on a 100-year time scale, and it remains in the atmosphere, on average, a **little more than a century.**


### **Fluorinated gases**
 fluorinated gases trap substantially more heat. Indeed, the GWP for these gases can be in the **thousands to tens of thousands**, and they have long atmospheric lifetimes, in some cases lasting tens of thousands of years, their global warming impact Varies according to HFC type(**from 140 to 11700**).


 [Source_1](https://unfccc.int/process/transparency-and-reporting/greenhouse-gas-data/greenhouse-gas-data-unfccc/global-warming-potentials)
 [Source_2](https://www.nrdc.org/stories/greenhouse-effect-101#gases)



#### The Order of gases according to  riskiness:
1.   Fluorinated gases (riskiest)
2.   Nitrous oxide
3.   Methane
4.   Carbon dioxide

```python
import statsmodels.api as sm

df_filtered["Emission"] = df_filtered["Emission"] - df_filtered["Emission"].shift(1)

# Drop missing values created by differencing
df_filtered = df_filtered.dropna()
#print(df_filtered)
cubed_root_df = df_filtered["Emission"].values ** (1/3)
df_filtered["Emission"] = cubed_root_df
df_filtered = df_filtered.dropna()
adf_test = sm.tsa.stattools.adfuller(df_filtered["Emission"])
adf_statistic = adf_test[0]
p_value = adf_test[1]
critical_values = adf_test[4]

print(f'ADF Statistic: {adf_statistic}')
print(f'p-value: {p_value}')
```

```python
train_data = df_filtered[:]

train_data
```

```python
train_data = train_data.reset_index(drop=True)
```

```python
test_data = df_filtered[8:]
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
import numpy as np

def create_sequences(data, window_size):
    X, y = [], []

    #print(len(data), window_size)
    for i in range(len(data) - window_size):
        X.append(data[i:i+window_size])
        y.append(data[i+window_size])
    return np.array(X), np.array(y)
window_size= 2
x_train, y_train = create_sequences(train_data["Emission"], window_size)

x_test, y_test = create_sequences(test_data["Emission"], window_size)
```

```python
model = Sequential()
model.add(LSTM(units=50, input_shape=(window_size, 1)))
model.add(Dense(units=1))
model.compile(optimizer='adam', loss='mean_squared_error')
model.fit(x_train, y_train, epochs=500, batch_size=32)
```

```python
pred= model.predict(x_train )
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error

mean_squared_error(y_train, pred, squared=False)
```

#### **keeping emissions low in the future**
* Use less electricity
* Generate electricity without emissions
* Reduce household waste
    * Waste we put in landfills releases greenhouse gases. Almost half the gas released by landfill waste is methane, which is an especially potent greenhouse gas.
    * Conscious choices, including avoiding unnecessary purchases, buying secondhand, eliminating reliance on single-use containers, switching to reusable, and recycling, can all help reduce household waste

* decrease emissions from manufacturing
    * There are ways to decrease emissions from manufacturing. Using materials that aren’t made from fossil fuels and don’t release greenhouse gases is a good start. For example, bioplastics made from plants are an alternative to plastics that come from fossil fuels.
    * Companies can also use renewable energy sources to power factories and ship the products that they create in fuel-saving cargo ships.

* Take carbon dioxide out of the air
    * For example, planting trees, bamboo, and other plants increases the number of carbon sinks.
    * There are also carbon dioxide removal technologies, which may be able to pull large amounts of greenhouse gases out of the atmosphere.
